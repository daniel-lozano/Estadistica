```python
#NUMERO DE PASOS

N=10000

#marcaran si se dio un paso y con que valor se dio el paso
x=[]
y=[]
#marcaran la suma de los pasos en x y y 
px=[]
py=[]

#comenzara en (0,0)
x.append(0)
y.append(0)
```

```python
#escogiendo dimension
def dimension():
    a=random.rand()
    if (a<=0.5):
        return 1 # x
    return -1 # y

```

```python
for i in range(N):
    dim=dimension()
    lado=dimension()
    if(dim==1):
        x.append(lado)
        y.append(0)
        px.append(sum(x))
        py.append(sum(y))
    if(dim==-1):
        y.append(lado)
        x.append(0)
        px.append(sum(x))
        py.append(sum(y))
        
```

```python
def funcion():
    Radio=[]
    
    for i in range(N):
        dim=dimension()
        lado=dimension()
        if(dim==1):
            x.append(lado)
            y.append(0)
            px.append(sum(x))
            py.append(sum(y))
        if(dim==-1):
            y.append(lado)
            x.append(0)
            px.append(sum(x))
            py.append(sum(y))
        
    for i in range(N):
        Radio.append(sqrt(px[i]**2+py[i]**2))
    
    return Radio[-1]
```

```python
X=sum(x)
Y=sum(y)
plot(px,py,"--")
scatter(px[-1],py[-1],label="ultimo punto",color="r",s=50)

scatter(px[0],py[0],label="t=0 primer punto",color="g",s=50)
#xlim(-abs(X)-2,abs(X)+2)
#ylim(-abs(Y)-2,abs(Y)+2)
xlabel("$ x $",size=20)
ylabel("$ y $",size=20)

#legend()
```
