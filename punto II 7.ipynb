```python
#escogiendo dimension
def dimension():
    a=random.rand()
    if (a<=0.5):
        return 1 # x
    return -1 # y

```

```python
def funcion():
    Radio=[]
    x=[]
    y=[]
#marcaran la suma de los pasos en x y y 
    px=[]
    py=[]
    for i in range(N):
        dim=dimension()
        lado=dimension()
        if(dim==1):
            x.append(lado)
            y.append(0)
            px.append(sum(x))
            py.append(sum(y))
        if(dim==-1):
            y.append(lado)
            x.append(0)
            px.append(sum(x))
            py.append(sum(y))
        
    for i in range(N):
        Radio.append(sqrt(px[i]**2+py[i]**2))
    
    return Radio[-1]
```

```python
a=funcion()
print a
```

```python
M=100
N=100
Radios=[]
for i in range(M):
    Radios.append(funcion())
    
hist(Radios,50)
print len(Radios)
xlabel("distancia alcanzada en "+ str(N)+ " pasos con "+str(M)+" puntos",size=20)
title(str(M)+" puntos ")

savefig("histograma.png")
```
